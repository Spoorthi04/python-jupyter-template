### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (First subplot)
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
axs[0, 0].set_title("Line Chart")

# Bar chart (Second subplot)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")

# Scatter plot (Third subplot)
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
axs[1, 0].scatter(scatter_x, scatter_y, color='orange', marker='o')
axs[1, 0].set_title("Scatter Plot")

# Pie chart (Fourth subplot)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
